```python
import pandas as pd
import numpy as np
import datetime as dt
from WindPy import w
import alphalens
import factor_analysis as fa
%matplotlib inline
```

```python
w.start()
```

```python
today = dt.datetime.today().strftime('%Y%m%d')
start_date, end_date = '2015-12-31', '2020-9-30'
```

# 提取股票列表

```python
data = w.wset("sectorconstituent", "date=%s;windcode=CI005016.WI;field=wind_code,sec_name"%(today)) # 中信家电
```

```python
stocks = pd.DataFrame(data.Data, index=data.Fields)
stocks = stocks.T
```

```python
stocks.head()
```

# 提取价格数据

```python
data = w.wsd(stocks['wind_code'].tolist(), 'close', start_date, end_date, 'Period=M;Days=Alldays')
```

```python
prices = pd.DataFrame(data.Data, index=data.Codes, columns=pd.to_datetime(data.Times))
prices = prices.T
```

```python
prices.index.set_names(['date'], inplace=True)
```

### 去掉区间内上市的股票

```python
prices = prices.dropna(axis=1)
```

```python
prices.to_excel('../data/prices_monthly.xlsx')
```

# 生成因子数据

```python
lookahead_bias_months = 3
factor = prices.pct_change(lookahead_bias_months).shift(-lookahead_bias_months)
factor.to_excel('../data/factor_monthly.xlsx')
```

```python
factor = factor.stack()
```

```python
factor.index = factor.index.set_names(['date', 'asset'])
```

```python
factor_data = alphalens.utils.get_clean_factor_and_forward_returns(factor, prices, periods=(1, 2, 3))
# factor_data = fa.utils.get_clean_factor_and_forward_returns(factor, prices, periods=(1, 2, 3))
```
